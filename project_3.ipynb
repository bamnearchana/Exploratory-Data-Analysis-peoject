find all instance when "wind speed is 24 " and " visibility is 25"

```python
def wind_visibility(wind, visi):
    if wind >= 24 and visi == 25:
        return True
    else:
        return False
```

```python
df_filtered = df[df.apply(lambda row: wind_visibility(row['Wind Speed_km/h'], row['Visibility_km']), axis=1)]
```
