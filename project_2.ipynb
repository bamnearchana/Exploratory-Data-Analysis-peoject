<h1>find the correlation using pearson method</h1>

```python
df.drop(columns=['key','mode','explicit'], inplace=True)
```

```python
corr_df=df.corr(method='pearson',)
plt.figure(figsize=(14,8))
sns.heatmap(corr_df, annot=True, fmt='0.1f',cmap='inferno',
                    linecolor='Black', linestyle='--',linewidth=1)
plt.title('Correlation between variables')
```

```python
sample_df=df.sample(int(0.004*len(df)))
```

```python
len(sample_df)
```

<b>Regression plot between loudness and energy</b>

```python
plt.figure(figsize=(10,8))
sns.regplot(sample_df, y='loudness',x='energy',color='c', )
plt.title('regression plot( correlation) between loudness and energy')
```

<h1>'regression plot( correlation) between acousticness and popularity'</h2>

```python
plt.figure(figsize=(10,8))
sns.regplot(sample_df, x='acousticness',y='popularity',color='m', )
plt.title('regression plot( correlation) between acousticness and popularity')
```

```python
df['dates']=df.index.get_level_values('release_date')
```

```python
df['dates']=pd.to_datetime(df['dates'])
```

```python
years=df['dates'].dt.year
```

```python
sns.histplot(years)
```

<h2>bar chart year vs duration</h2>

```python
durations=df['duration']
```

```python
fig,ax=plt.subplots(figsize=(18,7))
sns.barplot(x=years,y=durations,ax=ax)
plt.title('Duration vs years')
plt.xticks(rotation=90)
plt.savefig('barchart_spotify_betwen_years.png')
```

<h2>line chart year vs duration</h2>

```python
fig,ax=plt.subplots(figsize=(18,7))
sns.lineplot(x=years,y=durations,ax=ax)
sns.set_style(style='whitegrid')
plt.title('Duration vs years')
plt.xticks(rotation=90)
plt.savefig('linechart_spotify_betwen_years.png')
```

```python
fig,ax=plt.subplots(figsize=(18,7))
sns.scatterplot(x=years,y=durations,ax=ax)
sns.set_style(style='whitegrid')
plt.title('Duration vs years')
plt.xticks(rotation=90)
plt.savefig('scatterchart_spotify_betwen_years.png')
```
