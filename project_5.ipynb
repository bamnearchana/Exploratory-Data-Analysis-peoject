# 1 Show the number of Confirmed, Deaths and Recovered cases in each Region.

```python
df.groupby('Region')['Confirmed','Deaths','Recovered'].sum()#Answer 1
```

# 2 Remove all the records where the Confirmed Cases is Less Than 10.

```python
df=df[df['Confirmed'] >=10]#answer 2
#df=df.reset_index(drop=True)
df
```

# 3 In which Region, maximum number of Confirmed cases were recorded ?

```python
max_confirmed = df['Confirmed'].max() #answers 3
df.loc[df['Confirmed']==max_confirmed, 'Region'].iloc[0]
```

# 4 In which Region, minimum number of Deaths cases were recorded ?

```python
min_confirmed = df['Deaths'].min() #answers 4
df.loc[df['Deaths']==min_confirmed, 'Region'].iloc[0]
```

# 5 How many Confirmed, Deaths & Recovered cases were reported from India till 29 April 2020 ?

```python
df[(df['Region'] == 'India') & (df['Date'] == '4/29/2020')]#Answer 5
```

# 6 Sort the entire data wrt No. of Confirmed cases in ascending order.

Sort the entire data wrt No. of Confirmed cases in ascending order.


```python
df=df.sort_values(by=['Confirmed'],ascending=True)#answer 
df.reset_index(drop=True)
```

# 6 Sort the entire data wrt No. of Recovered cases in descending order.

```python
df=df.sort_values(by=['Recovered'],ascending=False)
df.reset_index(drop=True)
```
