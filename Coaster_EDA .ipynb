# data preparation

.Dropping irrelevant columns and rows<br/>

```python
df=df.iloc[:, [0, 3, 4, 7, 41, 42, 43, 44, 45,50, 53, 54, 55]]
```

```python
df.dtypes
```

```python
df.shape
```

```python
#renaming the columns name
df=df.rename(columns={'coaster_name':'Coaster_Name',
                     'year_introduced':'Year_Introduced',
                     'opening_date_clean':'Opening_Date',
                     'speed_mph':'Speed_mph',
                     "height_ft":"Height_ft",
                     "Inversions_clean":"Inversions",
                     "Gforce_clean":"Gforce"})
```

```python
df.columns
```

```python
#Checking the null value
df.isnull().sum()
```

```python
#checking duplicates
df.duplicated()
```

```python
df=df.loc[~df.duplicated(subset=['Coaster_Name','Location','Opening_Date'])]\
    .reset_index(drop=True).copy()
```

```python
df= df.reset_index(drop=True).copy()
```

```python
sns.pairplot(df,vars=['Year_Introduced','Speed_mph', 'Height_ft','Inversions', 'Gforce']
            , hue='Type_Main')
#plt.savefig('pairplot_coaster.png')
```

```python
df_corr= df[['Year_Introduced','Speed_mph', 'Height_ft','Inversions',
    'Gforce']].dropna().corr()
```

```python
sns.heatmap(df_corr, annot=True, fmt='0.1f',cmap='viridis')
plt.savefig('heatmap_coaster.png')
```

```python
df['Location'].value_counts()
```

```python
df.query("Location != 'Other'").groupby('Location').size()
```

```python
df_query=df.query("Location != 'Other'").groupby('Location')['Speed_mph'].agg(['mean','count'])
```

```python
df_query_2=df_query.query('count>=10')
```

```python
df_query_2.sort_values('mean').plot(kind='barh')
```

```python
plt.figure(figsize=(12,5))
df_query_2.sort_values('mean')['mean'].plot(kind='barh')
plt.title('Average Coast Speed by Location')
plt.xlabel('Location')
plt.ylabel('Average Coast Speed')
```
